# M05: Conditional Statements

These cases practice using `if`, `else if`, and `else` to make decisions based on user input. Conditions are evaluated top-down — the first matching branch executes and the rest are skipped.

---

## Case 1: Positive or Negative Check

Reads an integer from the user and checks whether it is positive or negative using a simple `if-else`.

> 💡 Note that `>= 0` means zero is classified as positive. Adjust to `> 0` if you want to handle zero separately.

```java
import java.util.Scanner;

Scanner obj = new Scanner(System.in);
System.out.print("Enter a number: ");
int number = obj.nextInt();

if (number >= 0){
  System.out.println("The number is positive.");
}
else {
  System.out.println("The number is negative.");
}
```

## Case 2: Odd or Even Check

Uses the **modulo operator** `%` to determine if a number is odd or even. `number % 2` returns the remainder when divided by 2 — if it's `0`, the number is even; otherwise it's odd.

```java
Scanner obj = new Scanner(System.in);
System.out.print("Enter a number: ");
int number = obj.nextInt();

if (number % 2 == 0){
  System.out.println("The number is even.");
}
else {
  System.out.println("The number is odd.");
}
```

## Case 3: Passing Grade Check

Takes three `float` scores, computes the total and average, then checks if the student passed or failed.

- `float` is used instead of `int` to handle decimal scores
- Passing threshold: **average ≥ 70**

```java
Scanner obj = new Scanner(System.in);
System.out.print("Score 1: ");
float num1 = obj.nextFloat();
System.out.print("Score 2: ");
float num2 = obj.nextFloat();
System.out.print("Score 3: ");
float num3 = obj.nextFloat();
float Total = num1 + num2 + num3;
System.out.println("Total = " + Total);

float Avg = Total / 3;
System.out.println("Avg = " + Avg);

if (Avg >= 70){
  System.out.println("Congratulations! You PASSED.");
}
else {
  System.out.println("You have failed. Please study hard.");
}
```

## Case 4: Menu Selection

Displays a menu and maps a user-entered item code to its name using `if-else if-else`.

- `String` comparison uses `.equals()` instead of `==` because `==` checks reference identity, not value equality
- Both uppercase and lowercase codes are accepted using `||` (OR)
- The `else` branch acts as a catch-all for invalid inputs
- Text block (`"""..."""`), introduced in Java 13, is used here to print the formatted menu cleanly

```java
Scanner obj = new Scanner(System.in);
System.out.println("""
Item Code       Item Name \n
A               Apple\n
B               Banana\n
C               Cherry""");

System.out.print("Item Code: ");
String code = obj.next();

if (code.equals("A") || code.equals("a")){
  System.out.println("Item Name = Apple");
}
else if (code.equals("B") || code.equals("b")){
  System.out.println("Item Name = Banana");
}
else if (code.equals("C") || code.equals("c")){
  System.out.println("Item Name = Cherry");
}
else {
  System.out.println("Unsupported code :(");
}
```

## Case 5: Year Level Identifier

Accepts a student's name and year level code, then prints a personalized greeting with their year level spelled out.

| Code | Year Level |
|------|------------|
| F / f | Freshman |
| S / s | Sophomore |
| J / j | Junior |
| anything else | `[undefined]` |

> ⚠️ The `G` (Graduate) code shown in the prompt is not handled — it falls into the `[undefined]` else branch. This could be added as another `else if`.

```java
Scanner obj = new Scanner(System.in);
System.out.print("Enter firstname: ");
String fn = obj.next();
System.out.print("Enter lastname: ");
String sn = obj.next();
System.out.print("Enter year level (F/S/J/G): ");
String yr = obj.next();

if (yr.equals("F") || yr.equals("f")){
  System.out.println("Hi " + fn + " " + sn + ", a freshman student.");
}
else if (yr.equals("S") || yr.equals("s")){
  System.out.println("Hi " + fn + " " + sn + ", a sophomore student.");
}
else if (yr.equals("J") || yr.equals("j")){
  System.out.println("Hi " + fn + " " + sn + ", a junior student.");
}
else {
  System.out.println("Hi " + fn + " " + sn + ", a [undefined] student.");
}
```

## Case 6: Grade Rating and Remarks

Extends Case 3 by assigning a numerical rating and remark based on the average score.

| Average | Rate | Remark |
|---------|------|--------|
| ≥ 90 | 4.0 | Excellent |
| ≥ 80 | 3.0 | Satisfactory |
| ≥ 70 | 2.0 | Passed |
| < 70 | 0.5 | Failed |

> ⚠️ **Bug in code:** The third condition is `Avg >= 80` (duplicate of the second), so the "Passed" branch (rate 2.0) is **unreachable**. It should be `Avg >= 70` to work correctly.

```java
Scanner obj = new Scanner(System.in);
System.out.print("Score 1: ");
float num1 = obj.nextFloat();
System.out.print("Score 2: ");
float num2 = obj.nextFloat();
System.out.print("Score 3: ");
float num3 = obj.nextFloat();
float Total = num1 + num2 + num3;
System.out.println("Total = " + Total);

float Avg = Total / 3;
System.out.println("Avg = " + Avg);
String rate;
String remark;

if (Avg >= 90){
  rate = "4.0";
  remark = "Excellent";
}
else if (Avg >= 80){
  rate = "3.0";
  remark = "Satisfactory";
}
else if (Avg >= 80){
  rate = "2.0";
  remark = "Passed";
}
else {
  rate = "0.5";
  remark = "Failed";
}

System.out.println("Rate: " + rate);
System.out.println("Remark: " + remark);

```

## Case 7: Menu with Size-Based Pricing

Extends Case 4 by adding a **nested `if`** inside each item branch to determine price based on size (LG or MED). This demonstrates how conditionals can be layered — the outer `if` checks the item code, and the inner `if` checks the size.

| Item | LG Price | MED Price |
|------|----------|-----------|
| Apple (A) | ₱100 | ₱50 |
| Banana (B) | ₱200 | ₱100 |
| Cherry (C) | ₱500 | ₱400 |

> 💡 `price` is initialized to `0` before the conditionals so it always has a value, even if an invalid code is entered.

```java
Scanner obj = new Scanner(System.in);
System.out.println("""
Item Code       Item Name     Size\n
                              LG    MED
A               Apple         100   50\n
B               Banana        200   100\n
C               Cherry        500   400""");

System.out.print("Item Code: ");
String code = obj.next();
System.out.print("Size: ");
String size = obj.next();
float price = 0;


if (code.equals("A") || code.equals("a")){
  System.out.println("Item Name = Apple");
  if (size.equals("LG") || size.equals("lg")){
     price = 100;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 50;
  }
}
else if (code.equals("B") || code.equals("b")){
  System.out.println("Item Name = Banana");
  if (size.equals("LG") || size.equals("lg")){
    price = 200;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 100;
  }
}
else if (code.equals("C") || code.equals("c")){
  System.out.println("Item Name = Cherry");
  if (size.equals("LG") || size.equals("lg")){
    price = 500;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 400;
  }
}
else {
  System.out.println("Unsupported code :(");
}

System.out.println("Price: PHP " + price);
```

## Case 8: Quantity and Discount

Extends Case 7 by accepting a **quantity** and applying a **discount** based on the total amount before checkout.

- Total amount = `price × qty`
- Discount rules:
  - Amount **≥ ₱1,000** → **10% off** (multiplied by `0.9`)
  - Amount **< ₱1,000** → **5% off** (multiplied by `0.95`)

> 💡 `(float)0.9` is an explicit cast to avoid Java treating the literal as a `double`, ensuring type consistency with the `float amt` variable.

```java
Scanner obj = new Scanner(System.in);
System.out.println("""
Item Code       Item Name     Size\n
                              LG    MED
A               Apple         100   50\n
B               Banana        200   100\n
C               Cherry        500   400""");

System.out.print("Item Code: ");
String code = obj.next();
System.out.print("Size: ");
String size = obj.next();
System.out.print("Qty: ");
int qty = obj.nextInt();
float price = 0;


if (code.equals("A") || code.equals("a")){
  System.out.println("Item Name = Apple");
  if (size.equals("LG") || size.equals("lg")){
     price = 100;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 50;
  }
}
else if (code.equals("B") || code.equals("b")){
  System.out.println("Item Name = Banana");
  if (size.equals("LG") || size.equals("lg")){
    price = 200;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 100;
  }
}
else if (code.equals("C") || code.equals("c")){
  System.out.println("Item Name = Cherry");
  if (size.equals("LG") || size.equals("lg")){
    price = 500;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 400;
  }
}
else {
  System.out.println("Unsupported code :(");
}

float amt = qty * price;
if (amt >= 1000){
  amt = amt * (float)0.9;
} else {
  amt = amt * (float)0.95;
}
System.out.println("Price: PHP " + amt);
```

## Case 9: Payment and Change

Extends Case 8 by accepting a **payment amount** and computing the change. Adds one more conditional layer:

- If payment is **less than** the discounted total → print "Insufficient Payment" and return the payment as-is
- Otherwise → compute `change = payment - amt`

This case combines item lookup, size pricing, quantity, discount, and payment validation into a single complete checkout flow.

```java
Scanner obj = new Scanner(System.in);
System.out.println("""
Item Code       Item Name     Size\n
                              LG    MED
A               Apple         100   50\n
B               Banana        200   100\n
C               Cherry        500   400""");

System.out.print("Item Code: ");
String code = obj.next();
System.out.print("Size: ");
String size = obj.next();
System.out.print("Qty: ");
int qty = obj.nextInt();
float price = 0;


if (code.equals("A") || code.equals("a")){
  System.out.println("Item Name = Apple");
  if (size.equals("LG") || size.equals("lg")){
     price = 100;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 50;
  }
}
else if (code.equals("B") || code.equals("b")){
  System.out.println("Item Name = Banana");
  if (size.equals("LG") || size.equals("lg")){
    price = 200;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 100;
  }
}
else if (code.equals("C") || code.equals("c")){
  System.out.println("Item Name = Cherry");
  if (size.equals("LG") || size.equals("lg")){
    price = 500;
  }  else if (size.equals("MED") || size.equals("med")){
    price = 400;
  }
}
else {
  System.out.println("Unsupported code :(");
}

float amt = qty * price;
if (amt >= 1000){
  amt = amt * (float)0.9;
} else {
  amt = amt * (float)0.95;
}
System.out.println("Price: PHP " + amt);

System.out.print("Payment: ");
float payment = obj.nextFloat();
float change = 0;
if (payment < amt){
  System.out.println("Insufficient Payment :(");
  change = payment;
} else{
  change = payment - amt;
}
System.out.println("Change: PHP " + change);
```
