# Welcome to My Java Notebook

## How to Use
1. Select a notebook from the file browser on the left
2. Each notebook consists of two types of cells:
   - **Markdown cells** — provide explanations and context
   - **Code cells** — contain runnable Java code
3. To run a code cell, click inside it and press **Shift + Enter**, or click the **▶** button in the toolbar
<img src="https://i.imgur.com/qS3tBfY.png" alt="run button location" width="50%"/>
4. If it doesn't run, check if the proper kernel is used. The inidicator on the top right should be `Java` and the status`idle` (empty circle).
<img src="https://i.imgur.com/uJFvHpg.png" alt="kernel button location" width="50%"/>
5. Code in cells is written without boilerplate — no `main` method or class wrapper. When transferring to NetBeans, place the code inside your `main` method.
6. Classes and methods defined in earlier cells must be run first before they can be used in later cells.
