## Case 1
Temp Conversion

```java
import java.util.Scanner;

// Case 1:
Scanner obj = new Scanner(System.in);
System.out.print("Enter C = ");
double C_in = obj.nextDouble();
System.out.println("ANSWER F = " + (C_in * 9 / 5 + 32) + "° F");
System.out.print("Enter F = ");
double F_in = obj.nextDouble();
System.out.println("ANSWER C = " + ((F_in - 32) * 5 / 9) + "° C");
```

## Case 2:
area calc

```java
import java.util.Scanner;
import java.lang.Math;

// Area Square
Scanner obj = new Scanner(System.in);
System.out.println("***AREA SQUARE***");
System.out.print("Side Length (m) = ");
double side_in = obj.nextDouble();
System.out.println("Area = " + (Math.pow(side_in, 2)) + " m²");
// Area Circle
System.out.println("***AREA CICLE***");
System.out.print("Radius (m) = ");
double rad_in = obj.nextDouble();
System.out.println("Area = " + (Math.pow(rad_in, 2) * Math.PI) + " m²");

// Area Triangle
System.out.println("***AREA TRIANGLE***");
System.out.print("Base Length (m) = ");
double tribase_in = obj.nextDouble();
System.out.print("Height (m) = ");
double triheight_in = obj.nextDouble();
System.out.println("Area = " + ((tribase_in * triheight_in) / 2) + " m²");

// Area Rectangle
System.out.println("***AREA Rectangle***");
System.out.print("Width (m) = ");
double width_in = obj.nextDouble();
System.out.print("Height (m) = ");
double recheight_in = obj.nextDouble();
System.out.println("Area = " + (width_in * recheight_in) + " m²");

// Area Trapezoid
System.out.println("***AREA Trapezoid***");
System.out.print("Major Base Length (m) = ");
double majorbase_in = obj.nextDouble();
System.out.print("Minor Base Length (m) = ");
double minorbase_in = obj.nextDouble();
System.out.print("Height (m) = ");
double trapheight_in = obj.nextDouble();
System.out.println("Area = " + (trapheight_in * (majorbase_in + minorbase_in) / 2 ) + " m²");

```
