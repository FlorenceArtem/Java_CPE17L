## Java Boilerplate

Every Java program needs a class and a `main` method as its entry point.
- `public` means it's accessible from outside the class
- `static` means `main` can be called without creating an instance of the class
- `String[] args` holds any command-line arguments passed to the program

```java
public class Sample {
    public static void main(String[] args) {
        // TODO code application logic here
    }
}
```

## Data Types and Print Function

Java is **statically typed** — you must declare a variable's type before using it. Common types include:

| Type | Example | Description |
|------|---------|-------------|
| `int` | `100` | Whole numbers |
| `double` | `107.75` | Decimal numbers |
| `char` | `'A'` | Single character (single quotes) |
| `String` | `"Anne"` | Text/sequence of characters (double quotes) |

`System.out.println()` prints output followed by a newline, so each subsequent print starts on a new line. Use `+` to concatenate variables with strings.

```java
int num1 = 100;
String name = "Anne";
double money = 107.75;
char letter = 'A';

System.out.println("The number is " + num1);
System.out.println("Hi " + name + " Have a great day!");
System.out.println("The amount is only " + money);
System.out.println("The letter is " + letter);
```

## Arithmetic Operations

Java supports standard arithmetic operators: `+` (add), `-` (subtract), `*` (multiply), `/` (divide).

> ⚠️ **Note:** Dividing two `int` values performs **integer division** — the decimal part is truncated. Wrap expressions in parentheses when using them inside string concatenation.

```java
int num1 = 100;
int num2 = 90;
int num3 = 80;

System.out.println("Quiz 1 = " + num1);
System.out.println("Quiz 2 = " + num2);
System.out.println("Quiz 3 = " + num3);
System.out.println("Total = " + (num1 + num2 + num3));
System.out.println("Avg = " + ((num1 + num2 + num3) / 3));
```

## User Input

To accept input from the user, Java uses the `Scanner` class from `java.util`. You must:
1. **Import** the class at the top: `import java.util.Scanner;`
2. **Create** a `Scanner` object tied to `System.in` (the keyboard)
3. **Read** input using methods like `nextInt()`, `nextDouble()`, or `nextLine()` depending on the expected data type. By default `next()` just gets strings separated by spaces.

```java
import java.util.Scanner;

Scanner obj = new Scanner(System.in);

System.out.print("Num1 = ");
int num1 = obj.nextInt();

System.out.print("Num2 = ");
int num2 = obj.nextInt();

System.out.println("The sum is " + (num1 + num2));
System.out.println("The diff is " + (num1 - num2));
System.out.println("The product is " + (num1 * num2));
System.out.println("The quotient is " + (num1 / num2));
```
