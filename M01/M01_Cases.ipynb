## Case 1: Basic Arithmetic with Fixed Values

Given two hardcoded integers, perform and display the four basic arithmetic operations: addition, subtraction, multiplication, and division.

**Given:** `num1 = 100`, `num2 = 10`

**Expected Output:**
```
The sum is 110
The diff is 90
The product is 1000
The quotient is 10
```

````java
int num1 = 100;
int num2 = 10;

System.out.println("The sum is " + (num1 + num2));
System.out.println("The diff is " + (num1 - num2));
System.out.println("The product is " + (num1 * num2));
System.out.println("The quotient is " + (num1 / num2));
````

## Case 2: Quiz Score Average

Given three quiz scores, display each score along with the total and average. Note that since all values are `int`, the average uses **integer division** (decimal truncated).

**Given:** `num1 = 100`, `num2 = 90`, `num3 = 80`

**Expected Output:**
```
Quiz 1 = 100
Quiz 2 = 90
Quiz 3 = 80
Total = 270
Avg = 90
```

````java
int num1 = 100;
int num2 = 90;
int num3 = 80;

System.out.println("Quiz 1 = " + num1);
System.out.println("Quiz 2 = " + num2);
System.out.println("Quiz 3 = " + num3);
System.out.println("Total = " + (num1 + num2 + num3));
System.out.println("Avg = " + ((num1 + num2 + num3) / 3));
````

## Case 3: Arithmetic with User Input

Instead of hardcoded values, this case accepts two integers from the user at runtime using `Scanner`. The same four arithmetic operations are then performed on the input values.

> ⚠️ `System.out.print()` (without `ln`) is used here so the input cursor stays on the same line as the prompt.

````java
import java.util.Scanner;

Scanner obj = new Scanner(System.in);

System.out.print("Num1 = ");
int num1 = obj.nextInt();

System.out.print("Num2 = ");
int num2 = obj.nextInt();

System.out.println("The sum is " + (num1 + num2));
System.out.println("The diff is " + (num1 - num2));
System.out.println("The product is " + (num1 * num2));
System.out.println("The quotient is " + (num1 / num2));
````

## Case 5: Student Information Form

Collects multiple `String` fields from the user (name, student number, age, year level) using `Scanner.next()`, which reads one whitespace-delimited token at a time. The inputs are then formatted into a short introduction.

> 💡 `next()` reads a single word. If you need to read a full line with spaces (e.g., a full middle name), use `nextLine()` instead.

````java
import java.util.Scanner;

Scanner obj = new Scanner(System.in);

System.out.print("Last Name: ");
String surn = obj.next();

System.out.print("First Name: ");
String firstn = obj.next();

System.out.print("Middle Initial: ");
String mi = obj.next();

System.out.print("Student Number: ");
String studNo = obj.next();

System.out.print("Age: ");
String age = obj.next();

System.out.print("Year Level: ");
String yr = obj.next();

System.out.println("Hi! I am " + firstn + " " + mi + " " + surn + ".");
System.out.println("My student number is " + studNo);
System.out.println("I am " + age + " years old.");
System.out.println("I am a " + yr + " student.");
````
