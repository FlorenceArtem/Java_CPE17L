## Data Types and Print function
- static type so you need to declare it before
- `System.out.println()` for newline terminated output (next print starts in next line)

````python
public class Sample {
    public static void main(String[] args) {
        int num1 = 100;
        String name = "Anne";
        double money = 107.75;
        char letter = 'A';
        System.out.println("The number is " + num1);
        System.out.println("Hi " + name + " Have a great day!");
        System.out.println("The amount is only " + money);
        System.out.println("The letter is " + letter);
    }
}
````

## Case 1:
given: 100, 10

output:
```
The sum is 110
The diff is 90
The product is 1000
The quotient is 10
```

````python
public class Sample {
    public static void main(String[] args) {
        int num1 = 100;
        int num2 = 10;
        System.out.println("The sum is " + (num1 + num2));
        System.out.println("The diff is " + (num1 - num2));
        System.out.println("The product is " + (num1 * num2));
        System.out.println("The quotient is " + (num1 / num2));
    }
}
````

## Case 2:
100, 90, 80

Output:
```
Quiz 1 = 100
Quiz 2 = 90
Quiz 3 = 80
Total = 270
Avg = 90
```

````python
public class Sample {
    public static void main(String[] args) {
        int num1 = 100;
        int num2 = 90;
        int num3 = 80;
        System.out.println("Quiz 1 = " + num1);
        System.out.println("Quiz 2 = " + num2);
        System.out.println("Quiz 3 = " + num3);
        System.out.println("Total = " + (num1 + num2 + num3));
        System.out.println("Avg = " + ((num1 + num2 + num3) / 3));
    }
}
````
